Bar Chart for Booking for new flats

```
import numpy as np
import matplotlib.pyplot as plt
#	At least one bar chart
#	At least one line chart
#	At least one pie chart
#	At least one histogram 
#	At least one scatterplot
#	At least one boxplot 

### Read the csv file with the loadtxt() function
fname = "/content/bookings-for-new-flats.csv"
newflatbookings = np.loadtxt(fname, skiprows=1,dtype=[('financial_year','i8'), ('no_of_units','i8')],delimiter=",")

years = newflatbookings['financial_year']
units = newflatbookings['no_of_units']

plt.title("Bookings For New Flats")
plt.ylabel("Number Of Units")
plt.xlabel("Financial Year")

plt.bar(years, units)
```

Pie chart for No. of resale applications in 2020 Q2 By Flat Types


```
import numpy as np
import matplotlib.pyplot as plt
#	At least one pie chart

### Read the csv file with the loadtxt() function
fname = "/content/number-of-resale-applications-registered-by-flat-type.csv"
resaleapplications = np.loadtxt(fname, skiprows=1,dtype=[('quarter','U50'), ('flat_type','U50'), ('no_of_resale_applications','i8')],delimiter=",")

data_2020_Q2 = resaleapplications[resaleapplications['quarter'] == "2020-Q2"]

labels = data_2020_Q2['flat_type']
values = data_2020_Q2['no_of_resale_applications']

colors = ['#FFA500', '#2CCDBD', '#CD2C7D', '#057F4A', '#FF00FF', '#00FF12']

plt.figure(figsize=(10,10))
plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%')
```

Boxplot for Number Of Resale Applications Registered By Flat Type (2007 to 2020)

```
import numpy as np
import matplotlib.pyplot as plt
#	At least one boxplot

### Read the csv file with the loadtxt() function
fname = "/content/number-of-resale-applications-registered-by-flat-type.csv"
resaleapplications = np.loadtxt(fname, skiprows=1,dtype=[('quarter','U50'), ('flat_type','U50'), ('no_of_resale_applications','i8')],delimiter=",")

data_2room = resaleapplications[np.isin(resaleapplications['flat_type'], ['2-room'])]
data_2room = data_2room['no_of_resale_applications']

data_3room = resaleapplications[np.isin(resaleapplications['flat_type'], ['3-room'])]
data_3room = data_3room['no_of_resale_applications']

data_4room = resaleapplications[np.isin(resaleapplications['flat_type'], ['4-room'])]
data_4room = data_4room['no_of_resale_applications']

data_5room = resaleapplications[np.isin(resaleapplications['flat_type'], ['5-room'])]
data_5room = data_5room['no_of_resale_applications']

data_exec = resaleapplications[np.isin(resaleapplications['flat_type'], ['Executive'])]
data_exec = data_exec['no_of_resale_applications']

data_combined = np.empty([len(data_2room),5])
#Rearrange the data structure for boxplot
for i in range(len(data_2room)):
  data_combined[i] = np.array([data_2room[i],
                               data_3room[i],
                               data_4room[i],
                               data_5room[i],
                               data_exec[i]])
  
zone_labels = ['2 Room','3 Room','4 Room', '5 Room', 'Executive']

boxprops = dict(linestyle='-',linewidth = 2, color='blue')
medianprops = dict(linewidth=2, color='red')
flierprops = dict(marker='o',markerfacecolor = 'green', markersize=8)
plt.boxplot(data_combined, labels=zone_labels, boxprops=boxprops,medianprops=medianprops,flierprops=flierprops)

```
